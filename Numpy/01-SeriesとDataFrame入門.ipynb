# 1. Seriesオブジェクトの作成方法

## 1.1 NumPyのndarrayをpandasのSeriesオブジェクトに変換する

```python
# ライブラリの読み込み
import numpy as np
import pandas as pd
```

```python
# 1. ndarrayオブジェクトの生成
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(type(arr))    # <class 'numpy.ndarray'>
```

```python
# 2. 上で作成した ndarray オブジェクトを Series オブジェクトに変換
s1 = pd.Series(data=arr)
print(s1)
print(type(s1))     # <class 'pandas.core.series.Series'>
```

## 1.2 Pythonのリストを直接渡して Series オブジェクトを作成する

```python
# この方法で Series オブジェクトを作成可能ですが、行インデックスを指定していない場合はデフォルトで 0 ～ n になります
# s2 = pd.Series(data=['乔峰', '男', 33])
s2 = pd.Series(['乔峰', '男', 33])  # data 引数名は省略可能
print(s2)
print(type(s2))     # <class 'pandas.core.series.Series'>
```

## 1.3 Pythonのリストを渡して、Seriesオブジェクトを作成し、行インデックスを指定する

```python
# Seriesオブジェクトを作成する際に、インデックスを指定できる
myField = ['name', 'gender', 'age']
s3 = pd.Series(['乔峰', '男', 33], index=myField)  # data引数名は省略可能
# s3 = pd.Series(['乔峰', '男', 33], ['name', 'gender', 'age'])  # 引数名を省略してもOK
print(s3)
print(type(s3))     # <class 'pandas.core.series.Series'>
```

## 1.4 タプルを使ってSeriesオブジェクトを作成する

```python
s4 = pd.Series(('乔峰', '男', 33), index=['name', 'gender', 'age'])  # data引数名は省略可能
print(s4)
print(type(s4))     # Series型であることを確認
```

## 1.5 Pythonの辞書（dict型）を使ってSeriesオブジェクトを作成する

```python
# 辞書を使ってSeriesオブジェクトを作成する。辞書のキーは行ラベル（index）に、値はデータに対応する。
s5 = pd.Series({'name':'乔峰', 'gender':'男', 'age':33})        
print(s5) 
print(type(s5))     # Series型であることを確認
```

```python
s5.to_frame()
```

# 2. DataFrameオブジェクトを作成する

## 2.1 辞書を使ってDataFrameオブジェクトを作成する

```python
# 1. 辞書を作成してデータを記録する
dict_data = {
    'id': [1, 2, 3],
    'name': ['乔峰', '虚竹', '段誉'],
    'age': [33, 29, 21]
}

# 2. 上記の辞書を基にDataFrameオブジェクトを作成
# df1 = pd.DataFrame(dict_data)  
# df1 = pd.DataFrame(dict_data, index=['A', 'B', 'C'])   # 行ラベル（インデックス）を指定
# index: 行ラベルを指定、columns: 列名と順序を指定。存在しない列名の場合、その値は NaN になる
df1 = pd.DataFrame(dict_data, index=['A', 'B', 'C'], columns=['id', 'age', 'name'])   

# 3. 結果を表示
print(df1)
print(type(df1))    # DataFrame型であることを確認
```

## 2.2 リストとタプルを使ってDataFrameオブジェクトを作成する

```python
# 1. リストとタプルを基にデータセットを作成。1つのタプル = 1行のデータ
list_data = [(1, '乔峰', 33), (2, '虚竹', 29), (3, '段誉', 21)]

# 2. DataFrameオブジェクトを作成
# df2 = pd.DataFrame(list_data, index=['X', 'Y', 'Z'], columns=['id', 'age', 'name'])  # 行単位でデータを渡し、columnsで列名を指定
df2 = pd.DataFrame(list_data, index=['X', 'Y', 'Z'], columns=['id', 'name', 'age'])  # 行単位でデータを渡し、columnsで列名を指定

# 3. 結果を表示
print(df2)
print(type(df2))    # DataFrame型であることを確認
```

# 6. Series のブール値操作

```python
# 1. データソースを読み込み、DataFrame オブジェクトを取得
df = pd.read_csv('data/scientists.csv')
df
```

```python
# 2. ブール値リストを手動で指定してデータを取得
bool_list = [True, False, False, True, False, False, True, True]  # dfの行数と一致させる必要あり
df[bool_list]   # インデックス 0, 3, 6, 7 の行を取得
```

```python
# 3. 条件に基づくブール値でデータをフィルタリング
# 目的: 平均年齢より高い科学者を抽出する
# 3.1 年齢列を取得
# ages_series = df.Age
ages_series = df['Age']     # 明示的に列名で取得
ages_series
```

```python
# 3.2 平均年齢を計算
avg_age = ages_series.mean()
avg_age
```

```python
# 3.3 各年齢が平均年齢より大きいかを判定し、ブールリストを作成
print(ages_series < ages_series.mean())
df[ages_series < avg_age]
```

```python
# 3.4 平均年齢より高い科学者のデータを取得
ages_series[ages_series > avg_age]
```

```python
print(ages_series.map(lambda x:x**2))
```

# 7. Series の計算操作

## 7.1 Series と数値の演算、数値は Series の各要素に適用される

```python
 # ages_series + 10
ages_series * 2
```

## 7.2 2つの Series 間の演算、長さが同じ場合は対応する要素ごとに計算される

```python
ages_series + ages_series
```

## 7.3 2つの Series 間の演算、長さが異なる場合は対応する要素のみ計算され、不一致の要素は NaN で補完される

```python
new_series = ages_series.add(pd.Series([1, 10], index=['Alice', 'Bob']), fill_value=0)
# new_series
```

## 7.4 Series 同士の計算は、可能な限りインデックスに基づいて行われる。つまり、インデックスが一致するデータを優先して計算する
